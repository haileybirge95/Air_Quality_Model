```python
import ipywidgets as widgets
from IPython.display import display
import threading
import time
import random

# Define pollutant thresholds for air quality categories
def calculate_air_quality(pm25, pm10, no2, so2, co, temp, humidity, proximity, population):
    # Example logic incorporating new factors into air quality determination
    score = (
        (pm25 / 12) + (pm10 / 50) + (no2 / 53) + (so2 / 35) + (co / 4.4) +
        (temp / 40) + (humidity / 100) + (proximity / 20) + (population / 1000)
    )
    if score <= 5:
        return "Good"
    elif score <= 10:
        return "Moderate"
    elif score <= 15:
        return "Unhealthy"
    elif score <= 20:
        return "Very Unhealthy"
    else:
        return "Hazardous"

# Create sliders for pollutants and factors
pm25_slider = widgets.FloatSlider(value=10, min=0, max=200, step=0.1, description='PM2.5')
pm10_slider = widgets.FloatSlider(value=50, min=0, max=500, step=1, description='PM10')
no2_slider = widgets.FloatSlider(value=30, min=0, max=1000, step=1, description='NO2')
so2_slider = widgets.FloatSlider(value=20, min=0, max=500, step=1, description='SO2')
co_slider = widgets.FloatSlider(value=2, min=0, max=50, step=0.1, description='CO')
temp_slider = widgets.FloatSlider(value=25, min=-10, max=50, step=0.1, description='Temperature')
humidity_slider = widgets.FloatSlider(value=50, min=0, max=100, step=1, description='Humidity')
proximity_slider = widgets.FloatSlider(value=10, min=0, max=20, step=0.1, description='Proximity')
population_slider = widgets.FloatSlider(value=500, min=0, max=1000, step=10, description='Population')

# Display for Air Quality
air_quality_label = widgets.Label(value="Air Quality: Good")

# Variables to control the simulation
simulation_running = threading.Event()

# Function to simulate real-time air quality
def simulate_real_time():
    while simulation_running.is_set():
        # Randomly adjust pollutant levels
        pm25_slider.value = max(0, min(200, pm25_slider.value + random.uniform(-5, 5)))
        pm10_slider.value = max(0, min(500, pm10_slider.value + random.uniform(-10, 10)))
        no2_slider.value = max(0, min(1000, no2_slider.value + random.uniform(-20, 20)))
        so2_slider.value = max(0, min(500, so2_slider.value + random.uniform(-10, 10)))
        co_slider.value = max(0, min(50, co_slider.value + random.uniform(-0.5, 0.5)))
        temp_slider.value = max(-10, min(50, temp_slider.value + random.uniform(-1, 1)))
        humidity_slider.value = max(0, min(100, humidity_slider.value + random.uniform(-2, 2)))
        proximity_slider.value = max(0, min(20, proximity_slider.value + random.uniform(-0.5, 0.5)))
        population_slider.value = max(0, min(1000, population_slider.value + random.uniform(-20, 20)))
        time.sleep(1)

# Start simulation button
start_button = widgets.Button(description="Start Simulation")
stop_button = widgets.Button(description="Stop Simulation")

def start_simulation(change):
    if not simulation_running.is_set():
        simulation_running.set()
        threading.Thread(target=simulate_real_time, daemon=True).start()

def stop_simulation(change):
    simulation_running.clear()

start_button.on_click(start_simulation)
stop_button.on_click(stop_simulation)

# Display the sliders, air quality label, and buttons
display(widgets.VBox([
    pm25_slider, pm10_slider, no2_slider, so2_slider, co_slider,
    temp_slider, humidity_slider, proximity_slider, population_slider,
    air_quality_label, widgets.HBox([start_button, stop_button])
]))
```

```python
import pandas as pd
import streamlit as st

# Load the dataset
try:
    data_file = "updated_pollution_dataset.csv"  # Ensure this file is in the same directory
    pollution_data = pd.read_csv(data_file)
except FileNotFoundError:
    st.error("The dataset file 'updated_pollution_dataset.csv' was not found. Please ensure it is in the same directory.")

# If dataset is loaded successfully, proceed
if 'pollution_data' in locals():
    # Calculate initial values (mean) and ranges for each parameter
    initial_values = pollution_data.mean(numeric_only=True)
    parameter_ranges = {
        "PM2.5": (pollution_data["PM2.5"].min(), pollution_data["PM2.5"].max()),
        "PM10": (pollution_data["PM10"].min(), pollution_data["PM10"].max()),
        "NO2": (pollution_data["NO2"].min(), pollution_data["NO2"].max()),
        "SO2": (pollution_data["SO2"].min(), pollution_data["SO2"].max()),
        "CO": (pollution_data["CO"].min(), pollution_data["CO"].max()),
        "Temperature": (pollution_data["Temperature"].min(), pollution_data["Temperature"].max()),
        "Humidity": (pollution_data["Humidity"].min(), pollution_data["Humidity"].max()),
        "Proximity_to_Industrial_Areas": (pollution_data["Proximity_to_Industrial_Areas"].min(),
                                           pollution_data["Proximity_to_Industrial_Areas"].max()),
        "Population_Density": (pollution_data["Population_Density"].min(),
                               pollution_data["Population_Density"].max()),
    }

    # Define a function to calculate air quality
    def calculate_air_quality(pm25, pm10, no2, so2, co, temp, humidity, proximity, population):
        score = (
            (pm25 / 12) + (pm10 / 50) + (no2 / 53) + (so2 / 35) + (co / 4.4) +
            (temp / 40) + (humidity / 100) + (proximity / 20) + (population / 1000)
        )
        if score <= 5:
            return "Good"
        elif score <= 10:
            return "Moderate"
        elif score <= 15:
            return "Unhealthy"
        elif score <= 20:
            return "Very Unhealthy"
        else:
            return "Hazardous"

    # Streamlit app layout
    st.title("Air Quality Simulator")

    # Create sliders for each parameter
    pm25 = st.slider("PM2.5", min_value=float(parameter_ranges["PM2.5"][0]),
                     max_value=float(parameter_ranges["PM2.5"][1]), value=float(initial_values["PM2.5"]))

    pm10 = st.slider("PM10", min_value=float(parameter_ranges["PM10"][0]),
                     max_value=float(parameter_ranges["PM10"][1]), value=float(initial_values["PM10"]))

    no2 = st.slider("NO2", min_value=float(parameter_ranges["NO2"][0]),
                    max_value=float(parameter_ranges["NO2"][1]), value=float(initial_values["NO2"]))

    so2 = st.slider("SO2", min_value=float(parameter_ranges["SO2"][0]),
                    max_value=float(parameter_ranges["SO2"][1]), value=float(initial_values["SO2"]))

    co = st.slider("CO", min_value=float(parameter_ranges["CO"][0]),
                   max_value=float(parameter_ranges["CO"][1]), value=float(initial_values["CO"]))

    temp = st.slider("Temperature", min_value=float(parameter_ranges["Temperature"][0]),
                     max_value=float(parameter_ranges["Temperature"][1]), value=float(initial_values["Temperature"]))

    humidity = st.slider("Humidity", min_value=float(parameter_ranges["Humidity"][0]),
                         max_value=float(parameter_ranges["Humidity"][1]), value=float(initial_values["Humidity"]))

    proximity = st.slider("Proximity to Industrial Areas", min_value=float(parameter_ranges["Proximity_to_Industrial_Areas"][0]),
                           max_value=float(parameter_ranges["Proximity_to_Industrial_Areas"][1]), value=float(initial_values["Proximity_to_Industrial_Areas"]))

    population = st.slider("Population Density", min_value=float(parameter_ranges["Population_Density"][0]),
                            max_value=float(parameter_ranges["Population_Density"][1]), value=float(initial_values["Population_Density"]))

    # Calculate air quality dynamically
    air_quality = calculate_air_quality(pm25, pm10, no2, so2, co, temp, humidity, proximity, population)
    st.metric(label="Air Quality", value=air_quality)
```
